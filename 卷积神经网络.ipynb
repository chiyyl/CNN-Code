```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np

mnist = input_data.read_data_sets("d:/mnistdata", one_hot = True)

trainImg = mnist.train.images
trainLabel = mnist.train.labels

testImg = mnist.test.images
testLabel = mnist.test.labels

inputImage = tf.placeholder(tf.float32,shape=(None,784))
inputLabel = tf.placeholder(tf.float32,shape=(None,10))

step = [1,1,1,1]

def Accuarcy(targetData,targetLabel):
    global secResult
    calResult = Sess.run(secResult,feed_dict={inputImage:targetData,inputLabel:targetLabel})
    calMax = tf.argmax(calResult,axis=1)
    targetMax = tf.argmax(targetLabel,axis=1)
    equlNum = tf.reduce_sum(tf.cast(tf.equal(targetMax,calMax),dtype=tf.float32))
    AccuarcyResult = equlNum/targetLabel.shape[0]
    return Sess.run(AccuarcyResult)

    


#first conve
firImgData = tf.reshape(inputImage,[-1,28,28,1])
firWeight = tf.Variable(tf.truncated_normal([5,5,1,32],stddev=0.1))
firBais=tf.Variable(tf.zeros([32])+0.1)
firCon = tf.nn.conv2d(firImgData,filter=firWeight,strides=step,padding="SAME")
firConResult = firCon+firBais
fCon = tf.nn.relu(firConResult)
firPool = tf.nn.max_pool(fCon,[1,2,2,1],[1,2,2,1],padding="SAME")

#sec conve

secWeight = tf.Variable(tf.truncated_normal([5,5,32,64],stddev=0.1))
secBais = tf.Variable(tf.zeros([64])+0.1)
secCon = tf.nn.conv2d(firPool,filter=secWeight,strides=step,padding="SAME")
secConResult = secCon + secBais
secCon = tf.nn.relu(secConResult)
secPool = tf.nn.max_pool(secCon,[1,2,2,1],[1,2,2,1],padding="SAME")




kData = tf.reshape(secPool,[-1,7*7*64])

#first kn
firKnWeight = tf.Variable(tf.truncated_normal([7*7*64,1024],stddev=0.1))
firKnBais = tf.Variable(tf.zeros([1024])+0.1)

firKnMul = tf.matmul(kData,firKnWeight)
firKnADD = tf.add(firKnMul,firKnBais)
firstRle = tf.nn.relu(firKnADD)
firDropout = tf.nn.dropout(firstRle,0.6)


#sec kn
secKnWeight = tf.Variable(tf.truncated_normal([1024,10],stddev=0.1))
secKnBais = tf.Variable(tf.zeros([10]) +0.1)
secKnMul = tf.matmul(firstRle,secKnWeight)
tempSec = tf.add(secKnMul,secKnBais)
secResult = tf.nn.softmax(tempSec)


lossFun = tf.reduce_mean( -tf.reduce_sum(inputLabel*tf.log(secResult),reduction_indices=[1]))

# train_step = tf.train.GradientDescentOptimizer(0.1).minimize(lossFun)
train_step = tf.train.AdamOptimizer(1e-4).minimize(lossFun)

init= tf.global_variables_initializer()

with tf.Session() as Sess:
    Sess.run(init)
    
   
    
    for i in range(0,1000):
        batch_xs, batch_ys = mnist.train.next_batch(100)
        Sess.run(train_step,feed_dict={inputImage:batch_xs,inputLabel:batch_ys})
        if i%50==0:
            value = Accuarcy(targetData = testImg,targetLabel=testLabel)
            print(value)
        
print("end")

            
    





```
