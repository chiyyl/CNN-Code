```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np

mnist = input_data.read_data_sets("C:/Users/chiyy/Desktop/tensorflow code/mnistdata", one_hot = True)

trainImg = mnist.train.images
trainLabel = mnist.train.labels

testImg = mnist.test.images
testLabel = mnist.test.labels

inputImage = tf.placeholder(tf.float32,shape=(None,784))
inputLabel = tf.placeholder(tf.float32,shape=(None,10))

step = [1,1,1,1]

#first conve
firImgData = tf.reshape(inputImage,[-1,28,28,1])
firWeight = tf.Variable(tf.random_normal([5,5,1,32]))
firBais=tf.Variable(tf.zeros([32]))
firCon = tf.nn.conv2d(firImgData,filter=firWeight,strides=step,padding="SAME")
firConResult = firCon+firBais
firPool = tf.nn.max_pool(firConResult,[1,2,2,1],[1,2,2,1],padding="SAME")

#sec conve

secWeight = tf.Variable(tf.random_normal([5,5,32,64]))
secBais = tf.Variable(tf.zeros([64]))
secCon = tf.nn.conv2d(firPool,filter=secWeight,strides=step,padding="SAME")
secConResult = secCon + secBais
secPool = tf.nn.max_pool(secConResult,[1,2,2,1],[1,2,2,1],padding="SAME")




kData = tf.reshape(secPool,[-1,7*7*64])

#first kn
firKnWeight = tf.Variable(tf.random_normal([7*7*64,10]))
firKnBais = tf.Variable(tf.zeros([10]))

firKnMul = tf.matmul(kData,firKnWeight)
firKnADD = tf.add(firKnMul,firKnBais)
firstRle = tf.nn.relu(firKnMul)


#sec kn
secKnWeight = tf.Variable(tf.random_normal([10,10]))
secKnBais = tf.Variable(tf.zeros([10]))
secKnMul = tf.matmul(firstRle,secKnWeight)
secResult = tf.add(secKnMul,secKnBais)


lossFun = tf.reduce_mean( - tf.reduce_sum(tf.multiply(inputLabel,tf.log(secResult),1))) 

init= tf.global_variables_initializer()

with tf.Session() as Sess:
    Sess.run(init)
    lossResult = Sess.run(lossFun,feed_dict={inputImage:testImg,inputLabel:testLabel})
    print(lossResult)

    
    

#  检查 多个tensor输出代码
#     sR,fPool,sPool= Sess.run([secResult,firPool,secPool],feed_dict={inputImage:testImg,inputLabel:testLabel})
#     print("secResult shape {} firPool shape{}  secCon shape{}".format(sR.shape,fPool.shape,sPool.shape))








#first knn






```
